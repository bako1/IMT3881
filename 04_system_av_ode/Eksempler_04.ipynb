# System av ordin√¶re differensialligninger

## Eksplisitt metode

Betrakt fisk- og hai-systemet
$$
\begin{aligned}
F'(t) &= (2 - S(t)) F(t) \\
S'(t) &= (F(t) - 1) S(t)
\end{aligned}
$$
med $F(0) = F_0$ og $S(0) = S_0$

Eksplisitt skjema
$$
\begin{aligned}
F_{i + 1} &= F_i + \Delta t (2 - S_i) F_i \\
S_{i + 1} &= S_i + \Delta t (F_i - 1) S_i
\end{aligned}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Parametere

dt = 1/1000
F0 = 1.9
S0 = 0.1
T = 30

# Initialisering

N = int(T / dt + 1)
t = np.linspace(0, T, N)
F = np.zeros(np.shape(t))
S = np.zeros(np.shape(t))
F[0] = F0
S[0] = S0

# Beregne vha. skjemaet

for i in range(N - 1):
    F[i + 1] = F[i] + dt * (2 - S[i]) * F[i]
    S[i + 1] = S[i] + dt * (F[i] - 1) * S[i]

# Plot

plt.figure()
plt.plot(t, F, t, S)
plt.xlabel('t')
plt.ylabel('Populasjon av fisk og hai')
plt.legend(('fisk', 'hai'))
```

Med begrensede ressurser er systemet
$$
\begin{aligned}
F'(t) &= (2 - F(t) - S(t)) F(t) \\
S'(t) &= (F(t) - 1) S(t)
\end{aligned}
$$
med $F(0) = F_0$ og $S(0) = S_0$

Eksplisitt skjema
$$
\begin{aligned}
F_{i + 1} &= F_i + \Delta t (2 - F_i - S_i) F_i \\
S_{i + 1} &= S_i + \Delta t (F_i - 1) S_i
\end{aligned}
$$

```python
# Parametre

dt = 1/1000
F0 = 1.9
S0 = 0.1
T = 30

# Initialisering

N = int(T / dt + 1)
t = np.linspace(0, T, N)
F = np.zeros(np.shape(t))
S = np.zeros(np.shape(t))
F[0] = F0
S[0] = S0

# Beregne vha. skjemaet

for i in range(N - 1):
    F[i + 1] = F[i] + dt * (2 - F[i] - S[i]) * F[i]
    S[i + 1] = S[i] + dt * (F[i] - 1) * S[i]

# Plot

plt.figure()
plt.plot(t, F, t, S)
plt.xlabel('t')
plt.ylabel('Populasjon av fisk og hai')
plt.legend(('fisk', 'hai'))
```

Forenklet line√¶rt ligningsett
$$
\begin{aligned}
F'(t) &= 1 - S(t) \\
S'(t) &= F(t) - 1
\end{aligned}
$$
med $F(0) = F_0$ og $S(0) = S_0$

Eksplisitt skjema
$$
\begin{aligned}
F_{i + 1} &= F_i + \Delta t (1 - S_i) \\
S_{i + 1} &= S_i + \Delta t (F_i - 1)
\end{aligned}
$$

```python
# Parametre

dt = 1/1000
F0 = 1.9
S0 = 0.1
T = 20

# Initialisering

N = int(T / dt + 1)
t = np.linspace(0, T, N)
F = np.zeros(np.shape(t))
S = np.zeros(np.shape(t))
F[0] = F0
S[0] = S0

# Beregne vha. skjemaet

for i in range(N - 1):
    F[i + 1] = F[i] + dt * (1 - S[i])
    S[i + 1] = S[i] + dt * (F[i] - 1)

# Plot tidsserie

plt.plot(t, F, t, S)
plt.xlabel('t')
plt.ylabel('Populasjon av fisk og hai')
plt.legend(('fisk', 'hai'))

# Faseplot

plt.figure()
plt.plot(F, S)
plt.xlabel('F')
plt.ylabel('S')
plt.axis('equal')
```

## Implisitt metode

Implisitt skjema for den forenklede line√¶re ligningen over:
$$
\begin{aligned}
F_{i + 1} &= \frac{F_i + \Delta t - \Delta t (S_i - \Delta t)}{1 + \Delta t^2} \\
S_{i + 1} &= \frac{S_i - \Delta t + \Delta t (F_i + \Delta t)}{1 + \Delta t^2}
\end{aligned}
$$

```python
# Parametre

dt = 1/1000
F0 = 1.9
S0 = 0.1
T = 20

# Initialisering

N = int(T / dt + 1)
t = np.linspace(0, T, N)
F = np.zeros(np.shape(t))
S = np.zeros(np.shape(t))
F[0] = F0
S[0] = S0

# Beregne vha. skjemaet

for i in range(N - 1):
    F[i + 1] = (F[i] + dt - dt * (S[i] - dt)) / (1 + dt**2)
    S[i + 1] = (S[i] - dt + dt * (F[i] + dt)) / (1 + dt**2)

# Plot

plt.plot(t, F, t, S)
plt.xlabel('t')
plt.ylabel('Populasjon av fisk og hai')
plt.legend(('fisk', 'hai'))

plt.figure()
plt.plot(F, S)
plt.xlabel('F')
plt.ylabel('S')
plt.axis('equal')
```

## Crank‚ÄìNicolson-skjemaet

Crank‚ÄìNicolson-skjemaet for det forenklede line√¶re systemet over er
$$
\begin{aligned}
F_{i + 1} &= \frac{(1 - \frac14\Delta t^2) F_i + \Delta t(\frac12\Delta t + 1) - S_i \Delta t}{1 + \frac14\Delta t^2} \\
S_{i + 1} &= \frac{(1 - \frac14\Delta t^2) S_i + \Delta t(\frac12\Delta t - 1) + F_i \Delta t}{1 + \frac14\Delta t^2}
\end{aligned}
$$

```python
# Parametre

dt = 1/1000
F0 = 1.9
S0 = 0.1
T = 20

# Initialisering

N = int(T / dt + 1)
t = np.linspace(0, T, N)
F = np.zeros(np.shape(t))
S = np.zeros(np.shape(t))
F[0] = F0
S[0] = S0

# Beregne vha. skjemaet

for i in range(N - 1):
    F[i + 1] = ((1 - dt**2 / 4) * F[i] + dt * (dt / 2 + 1) -
                dt * S[i]) / (1 + dt**2 / 4)
    S[i + 1] = ((1 - dt**2 / 4) * S[i] + dt * (dt / 2 - 1) +
                dt * F[i]) / (1 + dt**2 / 4)

# Plot

plt.figure()
plt.plot(t, F, t, S)
plt.xlabel('t')
plt.ylabel('populasjon av fisk og hai')
plt.legend(('fisk', 'shai'))

plt.figure()
plt.plot(F, S)
plt.xlabel('F')
plt.ylabel('S')
plt.axis('equal')
```

## Scipy

Scipy-modulen har biblioteksfunksjoner for l√∏sning av systemer av differensialligninger.

```python
from scipy.integrate import odeint


def begrenset(y, t):
    """
    H√∏yresiden av ode-systemet for vekst med begrensede ressurser
    """
    return [(2 - y[0] - y[1]) * y[0],
            (y[0] - 1) * y[1]]

# Parametre

F0 = 1.9
S0 = 0.1
T = 20

# Beregn vha. scipy

t = np.linspace(0, T)
y = odeint(begrenset, [F0, S0], t)

# Plot

plt.figure()
plt.plot(t, y)
plt.xlabel('t')
plt.ylabel('Populasjon av fisk og hai')
plt.legend(('fisk', 'hai'))
```

## Eksempel: kast med luftmotstand

Ligningssystemet er 
$$
\begin{aligned}
\mathbf{x}'(t) &= \mathbf{v}(t) \\
\mathbf{v}'(t) &= \mathbf{g} - k|\mathbf{v}(t)|\mathbf{v}(t)
\end{aligned}
$$
med $v_x(0) = v_0\cos(\theta)$, $v_y(0) = v_0\sin(\theta)$ og $x(0) = y(0) = 0$

Eksplisitt skjema
$$
\begin{aligned}
x_{i + 1} &= x_i + \Delta t v_{x,i} \\
y_{i + 1} &= y_i + \Delta t v_{y,i} \\
v_{x, i + 1} &= v_{x, i} - \Delta t k \sqrt{v_{x,i}^2 + v_{y,i}^2} \, v_{x,i} \\
v_{y, i + 1} &= v_{y, i} - \Delta t \left( k \sqrt{v_{x,i}^2 + v_{y,i}^2} \, v_{y,i} + g \right)
\end{aligned}
$$

```python
# Parametre

v0 = 10
theta = np.pi / 4
g = 9.81

def eksplisitt(T, N, k=0):
    """
    L√∏s hele systemet for kast med luftmostand
    """
    dt = T / N

    x = np.zeros(N + 1)
    y = np.zeros(N + 1)
    vx = np.zeros(N + 1)
    vy = np.zeros(N + 1)

    x[0] = 0
    y[0] = 0
    vx[0] = v0 * np.cos(theta)
    vy[0] = v0 * np.sin(theta)

    for i in range(N):
        x[i + 1] = x[i] + dt * vx[i]
        y[i + 1] = y[i] + dt * vy[i]
        vx[i + 1] = vx[i] - dt * k * np.sqrt(vx[i]**2 + vy[i]**2) * vx[i]
        vy[i + 1] = vy[i] - dt * (k * np.sqrt(vx[i]**2 + vy[i]**2) * vy[i] + g)

    return x, y, vx, vy

# Plot for forskjellige verdier av luftmostandsparameteren

for k in np.linspace(0, .3, 5):
    x, y, vx, vy = eksplisitt(1.5, 100, k)
    plt.plot(x, y)
```
